## Python Conditions and If statements

```python
# If
a = 33
b = 200
if b > a:
  print("b is greater than a \n")

#Elif
a = 33
b = 33
if b > a:
  print("b is greater than a \n")
elif a == b:
  print("a and b are equal \n")

#Else
a = 200
b = 33
if b > a:
  print("b is greater than a \n")
elif a == b:
  print("a and b are equal \n")
else:
  print("a is greater than b \n")

#If Else
a = 200
b = 33
if b > a:
  print("b is greater than a \n")
else:
  print("b is not greater than a \n")

#Short Hand If
if a > b: print("a is greater than b \n")
    
#Ternary Operator
a = 2
b = 330
print("A") if a > b else print("B \n")

#Multiple Else
a = 330
b = 330
print("A") if a > b else print("= \n") if a == b else print("B")

#And
a = 200
b = 33
c = 500
if a > b and c > a:
  print("Both conditions are True \n")

#Or
a = 200
b = 33
c = 500
if a > b or a > c:
  print("At least one of the conditions is True \n")

#Nested If
x = 41

if x > 10:
  print("Above ten,")
  if x > 20:
    print("and also above 20!")
  else:
    print("but not above 20.")
```
